```python
import pandas as pd
df_mid1 = pd.read_excel("중간저장1.xlsx")  # 입력받을 파일명 

temp_mid1 = list(df_mid1["스타일명_중분류"])
print(temp_mid1)
for i in range(len(temp_mid1)) :
    temp_join_list = list(temp_mid1[i])
    if len(temp_join_list) != 1 :
          temp_mid1[i] = "_".join(temp_join_list)
            
temp_mid1
```

```python
import pandas as pd


"""
주피터 실행 파일과 같은 디렉토리에 엑셀을 넣어주세요
ex 중간저장1 파일은 본인의 파일명입니다.

"""

df = pd.read_excel("ATH_02.xlsx") # 입력받을 파일명 
li_style = df["파일명"].tolist()
ans_style = []
for i in li_style :
    j = i.replace(" `-- ", '')
    j = j.replace(" |-- ", '')
    j = j.replace("|      ", '')
    j = j.replace("|  ", '')
    j = j.replace('- ', '-')
    j = j.replace(' ', '-')
    j = j.replace('_', '-')
    j = j.replace('~$', '')
    

    temp_style = j.split("-")
    switch = 0
    for k in temp_style :
        try :
            ans_style.append(int(k))
            switch = 1
            break
        except :
            continue
    if switch == 0 :
        ans_style.append("null")
df["스타일번호"] = ans_style
df.head()
df.to_excel('중간저장1-2.xlsx', index=False) # 저장할 파일명 
```
