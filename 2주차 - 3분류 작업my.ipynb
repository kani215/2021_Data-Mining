# 1. 스타일명 중분류 자동입력 처리 

- 1. 뒤 값인 "~"의류 를불러오기 위해 "-" 와 "_" 기준으로 분할해주고
- 2. 뒤 값 뭉텅이를 가져온다
- 3. 사람이 한 번더 확인 후 전철

```python
import pandas as pd

"""
주피터 실행 파일과 같은 디렉토리에 엑셀을 넣어주세요
ex 중간저장1 파일은 본인의 파일명입니다.

"""
df = pd.read_excel("ATH_03.xlsx") # 입력받을 파일명 
li_temp = df["파일명"].tolist()

li = []
temp_last = []
index= []
for i in li_temp :
    temp = i.split("-")
    temp_last.append( temp[-1])

temp_last
    
df["스타일명_중분류"] = temp_last
df.head()
df.to_excel('중간저장1.xlsx', index=False) # 저장할 파일명 
```

# 2. 스타일명 대분류 

- 1. 중분류가 대다수 "~"의류 생태로 수작업 후

- 2. 맨 뒷값을 가져오는 식으로 우선 처리해준다.



```python
import pandas as pd
df_mid1 = pd.read_excel("중간저장1.xlsx").fillna("null")  # 입력받을 파일명 
temp_mid1 = list(df_mid1["스타일명_중분류"])

li_style_mid = []
for style_mid in temp_mid1 :
    temp_style_mid = style_mid.split()
    li_style_mid.append(temp_style_mid[-1])
li_style_mid

df_mid1["스타일명_대분류"] = li_style_mid
df_mid1.head()
df_mid1.to_excel('중간저장2.xlsx', index=False) # 저장할 파일명 
```
